```python
import numpy as np

list = [1,2,3,4,5]
print(list)
arr = np.array(list)
print(arr)
```

```python
# 문자열과 숫자가 혼합된 리스트 --> 문자열형으로 통일
arr = np.array([1, 'a', 3])
print(arr)
```

```python
# 정수와 실수가 혼합된 리스트 --> 실수형으로 통일
arr = np.array([1, 2, 0.3])
print(arr)
```

1차원과 2차원 arr의 shape 알아보기

```python
arr1 = np.array([1, 2, 3, 4, 5])
print(arr1)

arr2 = np.array([[1, 2],
                [3, 4],
                [5, 6]])
print(arr2)

print(arr1,shape)
print(arr2,shape)
```

```python
arr1 = np.array([2,4,6,8,10])
arr2 = np.array([5,10,15,20,25])
result = arr1 + arr2
print(result)
```

함수 where()을 활용한 필터링

```python
arr = np.array([38,16,1,27,55,8,19])
print(np.where(arr>20))
```

sort 함수는 배열을 정렬하는 데 사용
